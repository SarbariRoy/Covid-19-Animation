```python
dt = pd.read_csv("time-series-19-covid-combined.csv",parse_dates=['Date'])
dt['Country/Region'] = dt['Country/Region'].replace('United Kingdom', 'UK').replace('United Arab Emirates',"UAE").replace('Korea, South','South Korea')
```

```python
import numpy
from matplotlib.pylab import *
from mpl_toolkits.axes_grid1 import host_subplot
import matplotlib.animation as animation
import pandas as pd
import random 
import seaborn as sns
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
import matplotlib.animation as animation
from IPython.display import HTML
import datetime
from matplotlib.ticker import MaxNLocator
from mpl_toolkits.basemap import Basemap
# f0 = figure(num = 0, figsize = (12, 8))
```

```python
def updateData(current_date):
    ## map
    ax.clear()
    date = current_date
    m=Basemap(ax=ax, llcrnrlon=-180, llcrnrlat=-65,urcrnrlon=180,urcrnrlat=80)
    m.drawmapboundary(fill_color='white', linewidth=0)
    m.fillcontinents(color=continents, alpha=0.3)
    m.drawcoastlines(linewidth=0.1, color="white")

    data = dt[dt['Date'] == date]
    data['labels_enc'] = pd.factorize(data['Country/Region'])[0]
    ax.scatter(data['Long'], data['Lat'], s=data['Confirmed']/6, alpha=0.4, c=data['Deaths'], cmap=cmaps)
    ax.text( 170,- 58,'Total Confirmed cases of COVID-19\n\nData collected from Kaggle\nAnimation realized with Python and the matplotlib library by Sarbari Roy', ha='right', va='bottom', size=9, color='#555555' )
    ins = ax.inset_axes([0.01,0.1,0.15,0.4])
    
   
    ##### Line chart
    df_ax1 = tot[(tot['Date'] <= date)]
    ins.clear()
    ins.set_ylim([0,maxm])
    ins.set_xlim([tot['Date'].min(),tot['Date'].max()])
    ins.plot(df_ax1['Date'],df_ax1['Confirmed'], color = line_color, linewidth=2)
    dr = df_ax1['Deaths'].max()/df_ax1['Confirmed'].max()
    ins.get_xaxis().set_ticks([])
    ins.yaxis.set_ticks_position('right')
    ins.spines['right'].set_visible(False)
    ins.spines['top'].set_visible(False)   
    
    
    ## Barchart
    ax3.clear()
    dsn = df[df['Date'] == current_date].sort_values(['Confirmed'],ascending = False)
    snap = dsn[['Country/Region','Confirmed','Recovered',"Deaths","Open Cases",'Open Recovered']].head(12)
    ax3.barh(snap['Country/Region'],snap['Open Cases'],color = neutral)
    ax3.barh(snap['Country/Region'],snap['Recovered'],left = snap['Open Cases'].tolist(), color = positive)
    ax3.barh(snap['Country/Region'],snap['Deaths'],left = snap['Open Recovered'].tolist(), color = negative)
    ax3.xaxis.set_major_formatter(ticker.StrMethodFormatter('{x:,.0f}'))
    ax3.xaxis.set_ticks_position('top')
    ax3.tick_params(axis='x', colors=axis_color, labelsize=8)
    ax3.legend(('Open Cases', 'Recovered','Deaths'))
    ax3.set_axisbelow(True)
    
    
    
#     table
    tdr = data['Deaths'].max()/data['Confirmed'].max()
    ax4.clear()
    rt = df[df['Date'] <= date]
    table_data=[
    ["Countries Affected", len(rt['Country/Region'].unique())],
    ['Overall Death Rate',  str(round(tdr,2)*100) + "%"]
    ]
    colors = [["darkblue","w"],[ "azure","w"]]
    table = ax4.table(cellText=table_data , loc='center', cellColours=colors)
    table.set_fontsize(12)
    table.scale(1,2)
#     ax4.title('Loss by Disaster')
    table[(0, 1)].get_text().set_fontsize(12)
    table[(0, 1)].get_text().set_color('royalblue')
    table[(1, 0)].get_text().set_color('darkblue')
    table[(1, 1)].set_text_props(color = 'darkblue',fontsize = 'large')
    table[(0, 0)].get_text().set_color('white')
    ax4.text(1, 1.7, date.strftime('%d %B %Y'), transform=ax4.transAxes, 
             color='cornflowerblue', size=30, ha='right', weight=800)
    ax4.axis('off')
    
    
    #big table
    tab2 = dsn.head(25)    

    ax5.clear()
    tab2['Death Rate']= np.where((tab2.Confirmed>0) & (tab2.Deaths>0),round((tab2['Deaths']*100)/tab2['Confirmed'],2),0 )
    idx = (-tab2['Death Rate']).argsort()[:3]

    tab2['Death Rate'] = tab2['Death Rate'].apply(lambda x: str(x) + "%")
    tab2 = tab2[['Country/Region','Confirmed','Death Rate']]
    tab2.columns = ['Region', 'Confirmed',"Death Date"]
    cell_text = []
    for row in range(len(tab2)):
        cell_text.append(tab2.iloc[row])

    table2 = ax5.table(cellText=cell_text, colLabels=tab2.columns, loc='upper center')
    table2[(0,0)].set_facecolor('darkblue')
    table2[(0,0)].set_text_props(color = 'white')
    table2[(0,1)].set_facecolor('azure')
    table2[(0,2)].set_facecolor(negative)
    table2[(0,2)].set_text_props(color = 'white')
    
    for i in idx:
        table2[(i+1,2)].set_text_props(fontweight='bold', color = high_rate)
        
    table2.set_fontsize(12)
    table2.scale(.9,2)
    
    ax5.axis('off')
```

```python
import matplotlib
matplotlib.use('Qt5Agg')
my_dpi=96
barplot_color = 'maroon'
country_name = "silver"
axis_color = 'gray'
line_color = 'crimson'
face ='white'
dr_color ='firebrick'
cmaps = "coolwarm_r" #"coolwarm_r"#"autumn" #"gist_rainbow" #"Reds_r" #"PuRd_r"
neutral = 'silver' #'#cc5656' #'#a35050' #'#cad7e8'
positive = 'yellowgreen' #'#32a875'
negative = 'black'
continents = 'gray'
china = 'salmon'
high_rate = 'firebrick'


f0= figure(figsize=(2600/my_dpi, 1800/my_dpi), dpi=my_dpi,facecolor=face)
# f0= figure(figsize=(12,8), dpi=my_dpi)
f0.suptitle("Corona Virus Confirmed Cases Till Date", fontsize=26, weight=500, color = 'darkblue')
f0.tight_layout()
# plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)

# plt.style.use('classic')
date_ax  = subplot2grid((12, 16), (0, 13), colspan=3, rowspan=1) #tablw
ax4  = subplot2grid((12, 16), (0, 13), colspan=3, rowspan=1) #tablw
ax5 = subplot2grid((12, 16), (1, 13), colspan=3, rowspan=10) #table
ax  = subplot2grid((12, 16), (0, 0), colspan=12, rowspan=7) #map
ax3 = subplot2grid((12, 16), (8, 1), colspan=12, rowspan=4) #bar


clrs = sns.color_palette('RdBu', n_colors=len(dt['Country/Region'].unique()))
colors = dict(zip(
    dt['Country/Region'].unique(),
    clrs
))

#######
dt['check'] = dt['Country/Region'] +"#"+dt['Date'].apply(lambda x: str(x))
df = dt[['Country/Region', 'Lat', 'Long', 'Date', 'Confirmed', 'Deaths', 'Recovered']]
df['check'] = df['Country/Region'] +"#"+df['Date'].apply(lambda x: str(x))

df = df.groupby(['Country/Region', 'Date']).agg({'Confirmed': "sum",'Deaths': "sum",'Recovered':"sum"}).reset_index().sort_values(['Date'])
df = df[df['Confirmed'] > 0]
# df=df[df['Confirmed'] > 20]
df['check'] = df['Country/Region'] +"#"+df['Date'].apply(lambda x: str(x))

df['Open Cases'] = df['Confirmed'] -df['Deaths']- df['Recovered']
df['Open Recovered'] = df['Confirmed'] - df['Deaths']

##total_data
tot = df.groupby(['Date']).agg({'Confirmed': "sum",'Deaths': "sum",'Recovered':"sum"}).reset_index()
maxm = tot['Confirmed'].max()
#create animation
ranges = sorted(set(dt['Date'].tolist()), key=dt['Date'].tolist().index)
simulation = animation.FuncAnimation(f0, updateData, blit=False, frames=ranges, interval=1, repeat=False)

# Uncomment the next line if you want to save the animation
# simulation.save(filename='sim.mp4',fps=30,dpi=300)

plt.margins(0,0)
plt.subplots_adjust(top = .9, bottom = 0, right = .98, left = 0.02, 
            hspace = .1, wspace = 0.1)
plt.show()

# HTML(simulation.to_jshtml())
```
